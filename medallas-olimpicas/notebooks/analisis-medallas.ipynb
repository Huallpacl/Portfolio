```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv("../data/medallas.csv")
df.head()
```

```python
# Convertir columnas a enteros
for col in ["Oro", "Plata", "Bronce", "Total"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Revisar datos nulos
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

# Top 10 países por medallas totales
top10 = df.sort_values("Total", ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top10["País"], top10["Total"], color="gold")
plt.title("Top 10 países con más medallas")
plt.ylabel("Total de medallas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
